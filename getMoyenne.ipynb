```python
import pandas as pd
```

```python
df = pd.read_csv("./final_data/full_dep_data_to_2024.csv.gz", comment="#", sep=",")
df_station = pd.read_csv("./final_data/full_stations_metadata.csv", comment="#", sep=",")
df
```

```python
df_station = df_station.rename(columns={"num_poste" : "NUM_POST"})
df_station
```

```python
df['YYYY'] = df['YYYYMM'].astype(str).str[:4]
df['MM'] = df['YYYYMM'].astype(str).str[4:]

df['MM'] = df['MM'].astype(int)

df.drop('YYYYMM', axis=1, inplace=True)
df
```

```python
df_joined = df.merge(df_station, on="NUM_POST", how="left")
df_joined
```

```python
df_joined.dropna()
```

```python
grouped_df_year = df_joined.groupby(['num_dep', 'YYYY', 'TYPE']).agg({'VALEUR': 'mean'}).reset_index()
grouped_df_year
```

```python
grouped_df_month = df_joined.groupby(['num_dep', 'YYYY' ,'MM', 'TYPE']).agg({'VALEUR': 'mean'}).reset_index()
grouped_df_month
```

```python
grouped_df_year.to_csv("./final_data/groupedDataYear.csv", index=False)
grouped_df_month.to_csv("./final_data/groupedDataMonth.csv", index=False)
```
