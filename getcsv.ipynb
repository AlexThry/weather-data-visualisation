```python
import pandas as pd 
import os
```

```python
folderPath = './data/'
```

```python
filesFolder = os.listdir(folderPath)

cvsFiles = [file for file in filesFolder if file.endswith('.csv')]
dataFolders = [name for name in filesFolder if os.path.isdir(os.path.join(folderPath, name))]
print(cvsFiles)
print(dataFolders)
```

```python
dfreagrupped = []
for ind1, data in enumerate(cvsFiles):
    df = pd.read_csv(f"./data/{data}", sep=";", comment="#")
    csvNames = df["nom_fichier"].tolist()
    csvPoste = df["num_poste"].tolist()
    
    for ind2, namedata in enumerate(csvNames):
        df_data = pd.read_csv(f"./data/{dataFolders[ind1]}/{namedata}", sep=";", comment="#")
        df_data["TYPE"] = dataFolders[ind1].split("_")[1]
        df_data["NUM_POST"] = csvPoste[ind1]

        dfreagrupped.append(df_data)

        combined_df = pd.concat(dfreagrupped, ignore_index=True)
```
