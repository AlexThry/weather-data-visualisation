```python
import pandas as pd
import json

with open("stations.json") as f:
    data = json.load(f)

list_data = []

for i in list(data["features"].keys()):
    list_data.append(data["features"][i]["properties"])
    
df = pd.DataFrame(list_data)
df
```

```python
df.to_csv('stations_metadata.csv', index=False)
```
