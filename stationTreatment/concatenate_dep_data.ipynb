```python
import pandas as pd
import os
```

```python

dir = "../data/data_dep"
files = os.listdir(dir)

df_list = []

for file in files:
    df_list.append(pd.read_csv(os.path.join(dir, file), sep=";", comment="#"))

df = pd.concat(df_list, ignore_index=True)
df
    
```

```python
df_reduced = df[["AAAAMM", "TM", "TX", " TN", "RR", "NUM_POSTE"]].rename(columns={" TN": "TN"})
df_reduced
```

```python
df_final = df_reduced.melt(id_vars=['AAAAMM', 'NUM_POSTE'], 
                   value_vars=['TM', 'TX', 'TN', 'RR'], 
                   var_name='TYPE', 
                   value_name='VALEUR')
df_final = df_final.dropna().rename(columns={"AAAAMM": "YYYYMM", "NUM_POSTE": "NUM_POST"})
df_final
```

```python
df_in = pd.read_csv("../final_data/donnesIN.csv")
df_in
```

```python
df_export = pd.concat([df_final, df_in], ignore_index=True)
df_export
```

```python
df_export.to_csv("../final_data/full_dep_data_to_2024.csv.gz", compression='gzip', index=False)
```
