```python
import pandas as pd
import json

with open("stations.json") as f:
    data = json.load(f)

list_data = []

for i in list(data["features"].keys()):
    list_data.append(data["features"][i]["properties"])
    
df = pd.DataFrame(list_data)
df
```

```python
df.to_csv('../final_data/stations_metadata.csv', index=False)
```

```python
station_types = pd.read_csv("../final_data/postes_meteo.csv")
df['NUM_POSTE'] = df['NUM_POSTE'].astype(int)
merged_df = station_types.merge(df, left_on='num_poste', right_on='NUM_POSTE')
merged_df = merged_df.drop(["NUM_POSTE", "LAT_DG", "LON_DG", "ficheClimReduite", "ficheClimComplete"], axis=1)
merged_df = merged_df.rename(columns={'latitude(°)': 'latitude', 'longitude(°)': 'longitude', "COMMUNE": "commune", "ALTI": "altitude", "NOM_USUEL": "nom_usuel", "NUM_DEP": "num_dep"})
merged_df.head()
```

```python
merged_df.to_csv('../final_data/full_stations_metadata.csv', index=False)
```
